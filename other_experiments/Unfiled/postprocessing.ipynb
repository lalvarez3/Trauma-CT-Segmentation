```python
import skimage
import matplotlib.pyplot as plt


extrac = skimage.io.imread('M102.png')[:,:,:3]
im_label = skimage.measure.label(extrac )
```

```python
fig, ax = plt.subplots(1,2, figsize=(5,9))
ax[0].imshow(extrac)
ax[0].set_title('E. coli phase segmentation')
ax[1].imshow(im_label)
ax[1].set_title('B. subtilis phase segmentation')
plt.tight_layout()
```
